```python
data_tln = data[(data.Lon<25) & (data.Lon>24.58) & (data.Lat<59.54) & (data.Lat>59.36)]
```

```python
tln_lat_min = data_tln.Lat.min()
tln_lat_max = data_tln.Lat.max()
tln_lon_min = data_tln.Lon.min()
tln_lon_max = data_tln.Lon.max()
tln_lon_center = float((tln_lon_max + tln_lon_min) / 2)
tln_lat_center = float((tln_lat_max + tln_lat_min) / 2)
```

```python
map_ee = folium.Map(location=[tln_lat_center, tln_lon_center], zoom_start=11)
map_ee
```

```python
number_lats = 50
number_lons = 100
lat_step = float((tln_lat_max-tln_lat_min)/number_lats)
lon_step = float((tln_lon_max-tln_lon_min)/number_lons)
```

```python
%%time
region = 0
lat_1 = 0
lon_1 = 0
regions = pd.DataFrame(columns=['west','east','south','north','region'])
for lat in list(range(number_lats)):
    for lon in list(range(number_lons)):
        region += 1
        lat_0 = tln_lat_min + lat*lat_step
        lon_0 = tln_lon_min + lon*lon_step
        lat_1 = (tln_lat_min + (lat+1)*lat_step) if lat<number_lats else tln_lat_max
        lon_1 = (tln_lon_min + (lon+1)*lon_step) if lon<number_lons else tln_lon_max
        regions.loc[region-1] = [lon_0, lon_1, lat_0, lat_1, region]
regions.region = regions.region.astype(int)
```

```python
regions.to_csv('regions.csv')
```

```python
data_tln.loc[:,'region_id'] = 0
#data_tln.loc[:,'region_id'] = data_tln.loc[:,'region_id'].astype(int)
```

```python
%%time
regions_list=[]
for i in list(range(data_tln.shape[0])):
    lat = data_tln.iloc[i]['Lat']
    lon = data_tln.iloc[i,]['Lon']
    regions_list.append(regions[(regions.south<=lat) & ((regions.north>lat) | (lat==tln_lat_max)) &
                                (regions.west<=lon) & ((regions.east>lon) | (lon==tln_lon_max))].region.values[0])
data_tln.loc[:,'region_id'] = regions_list
```

```python
groupped = data_tln[['Kahju suurus (euro)','region_id']].groupby(['region_id']).sum().reset_index()
```

```python
def prepare_geo_data(data, thresh=0):

    geo_data_features = []

    # Отфильтруем ячейки с малым количеством поездок
    groupped2 = groupped[groupped['Kahju suurus (euro)'] >= thresh]

    log = np.log(groupped2['Kahju suurus (euro)'].values + 1)
    log = log / float(log.max())

    for i in range(groupped2.shape[0]):
        rid = groupped2.loc[i,'region_id']
        kahju_surus = log[i]
        point1 = [regions[regions.region==rid].west.values[0], regions[regions.region==rid].south.values[0]]
        point2 = [regions[regions.region==rid].west.values[0], regions[regions.region==rid].north.values[0]]
        point3 = [regions[regions.region==rid].east.values[0], regions[regions.region==rid].north.values[0]]
        point4 = [regions[regions.region==rid].east.values[0], regions[regions.region==rid].south.values[0]]

        geo_data_features.append({
            "properties": {"name": str(rid), "value": kahju_surus},
            "id": str(rid),
            "type": "Feature",
            "geometry": {
                "type": "Polygon",
                "coordinates": [[point1, point2, point3, point4]]
            }        
        })

    geo_data = {
        "type": "FeatureCollection",
        "features": geo_data_features
    }
    
    return geo_data
```

```python
geo_data = prepare_geo_data(data_tln, thresh=0)
```

```python
map_tln = folium.Map(location=[tln_lat_center, tln_lon_center], zoom_start=11)
folium.GeoJson(
    geo_data,
    style_function=lambda feature: {
        'color': '#330099',
        'weight': 0.0,
        'fillOpacity': 0.8 * feature['properties']['value'],
        'dashArray': None
    }).add_to(map_tln)

map_tln
```
